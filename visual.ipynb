```python

import torch
from FlagEmbedding.visual.modeling import Visualized_BGE

import matplotlib.pyplot as plt
from PIL import Image

def display_images_in_row(filenames):
    num_images = len(filenames)
    fig, axes = plt.subplots(1, num_images, figsize=(num_images * 2, 2))

    for i, filename in enumerate(filenames):
        img = Image.open(filename)
        img = img.resize((200, 200))  # Resize image to 100x100 pixels
        axes[i].imshow(img)
        axes[i].axis('off')  # Turn off axis

    plt.tight_layout()
    plt.show()

model = Visualized_BGE(model_name_bge = "BAAI/bge-base-en-v1.5", model_weight="./FlagEmbedding/Visualized_base_en_v1.5.pth")
model.eval()
```

```python
filenames = ['multi_modal_cases/case1/query.jpg', 'multi_modal_cases/case1/ref1.jpg', 'multi_modal_cases/case1/ref2.jpg']
display_images_in_row(filenames)
```

```python
with torch.no_grad():
    query_emb = model.encode(image="multi_modal_cases/case1/query.jpg", text="car in winter")
    candi_emb_1 = model.encode(image="multi_modal_cases/case1/ref1.jpg")
    candi_emb_2 = model.encode(image="multi_modal_cases/case1/ref2.jpg")

sim_1 = query_emb @ candi_emb_1.T
sim_2 = query_emb @ candi_emb_2.T
print(sim_1, sim_2)
```
